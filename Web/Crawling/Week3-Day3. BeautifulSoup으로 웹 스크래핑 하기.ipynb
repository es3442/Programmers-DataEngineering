# [Week3] Day3. BeautifulSoup 으로 웹 스크래핑 하기
- BeautifulSoup을 이용해서 Hashcode 질문 내용을 스크래핑해봅시다.

### Step1) BeautifulSoup 객체 만들기

```python
# 다음 User-Agent를 추가해봅시다.

user_agent = {"사용자 정보 입력"}
```

```python
# 필요한 라이브러리를 불러온 후, 요청을 진행해봅시다.
import requests
from bs4 import BeautifulSoup
res=requests.get("https://qna.programmers.co.kr/", user_agent)
```

```python
# 응답을 바탕으로 BeautifulSoup 객체를 생성해봅시다.
soup=BeautifulSoup(res.text, "html.parser")
```

### Step 2-1) 태그 이름으로 원하는 데이터 추출

```python
## id 없이 div 태그를 찾아봅시다.
soup.findAll("div")
```

### Step 2-2) 태크-id 이름으로 원하는 데이터 추출

```python
## id 없이 div 태그를 찾아봅시다.
soup.findAll("li", id="main-app-header")
```

### Step 2-2) 태크-Class 이름으로 원하는 데이터 추출

```python
## id 없이 div 태그를 찾아봅시다.
soup.findAll("li", "question-list-item")
```

### Step 3) 페이지네이션(Pagination)

페이지네이션은 많은 정보를 인덱스로 구분하는 기법

i번째 페이지

`https://hashcode.co.kr/?page={i}`

```python
# Pagination이 되어있는 질문 리스트의 제목을 모두 가져와봅시다.
# 과도한 요청을 방지하기 위해 1초마다 요청을 보내봅시다.
import time

for i in range(1, 6):
    res=requests.get("https://hashcode.co.kr/?page={}".format(i), user_agent)
    soup=BeautifulSoup(res.text, "html.parser")
    questions=soup.find_all("li", "question-list-item")
    for question in questions:
        print(question.find("div", "question").find("div", "top").h4.text.strip())

    time.sleep(0.5)
    
```
